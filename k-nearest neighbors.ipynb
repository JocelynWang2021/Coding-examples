## [Bayes Rule, 2 points]

Let’s say
<br>
Event A= Patient has this cancer 
<br>
Event B= Test shows positive result
<br>
<br>
Probability of having this cancer:
<br>
$P(A) = \frac{1}{12000}$
<br>
<br>
Probability of not having this cancer
<br>
$P(A^C) = 1 - P(A) = 1 -  \frac{1}{12000}$
<br>
<br>
Probability of testing positive when the patient has this cancer
<br>
P(B | A) = 0.992
<br>
<br>
Probability of testing negetive when the patient has this cancer
<br>
$P(B^C | A) = 1 - P(B | A) = 1 - 0.992 = 0.008$
<br>
<br>
Probability of testing negetive when the patient doesn't have this cancer
<br>
$P(B^C | A^C) = 0.992$
<br>
<br>
Probability of testing negetive when the patient doesn't have this cancer
<br>
$P(B | A^C) = 1 - P(B^C | A^C) = 0.008$
<br>
<br>
So according to our assumptions, We need to calculate P (A | B) Using Bayes Theorem we know:
<br>
$P(A | B)= \frac{P(B | A)*P(A)}{P(B)}$
<br>
<br>
And probability of having positive test:
<br>
$P(B) = P(B | A)*P(A) + P(B | A^C)*P(A^C)$
<br>
<br>
So:
<br>
$P(A | B) = \frac{P(B | A) * P(A)}{P(B | A)*P(A) + P(B | A^C)*P(A^C)}$
<br>
≈ 0.0102
<br>
So if the initial probability is only 1 in 12,000 people (0.0083%), a positive test will increase that to 1.02%.

## [Spherical Gaussian estimation, 9 points]

In this case, p = 2, 
Given data in form of a matrix 𝐗 of dimensions m×p, if we assume that the data follows a 𝑝-variate Gaussian distribution with parameters mean 𝜇 ( 𝑝×1) and covariance matrix Σ (𝑝×𝑝) the Maximum Likelihood Estimators are given by:
<br>
$\mu=\frac{1}{m}\sum_{i=1}^{m}𝐱^{(𝐢)}$
<br>
<br>
$𝜎^2 = Σ = \frac{1}{m}\sum_{i=1}^{m}(X^{(i)} - \mu)(X^{(i)} - \mu)^T $
<br>
<br>
So, $𝜎 = \sqrt{ Σ} = \sqrt{\frac{1}{p}\sum_{i=1}^{p}(X^{(i)} - \mu)(X^{(i)} - \mu)^T} $
<br>
<br>
Assume that we have 𝑚 random vectors, each of size 𝑝: $𝐗^{(1)}$,$𝐗^{(2)}$,...,$𝐗^{(m)}$ where each random vectors can be interpreted as an observation (data point) across 𝑝 variables. If each $𝐗^{(𝑖)}$ are i.i.d. as multivariate Gaussian vectors:
<br>
<br>
$𝐗^{(𝑖)}$ ~ $N(\mu, Σ)$ 
<br>
<br>
By the independence of the random vectors, the joint density of the data {$𝐗^{(𝐢)}$,𝑖=1,2,...,m} is the product of the individual densities, that is $\prod_{i=1}^{m}𝑓_{𝐗^{(𝐢)}}(𝐱^{(𝐢)};𝜇,𝚺)$. Taking the logarithm gives the log-likelihood function
<br>
<br>
$𝑙(𝜇,𝚺 | (𝐱^{(𝐢)})$ = log $\prod_{i=1}^{m}𝑓_{𝐗^{(𝐢)}}(𝐱^{(𝐢)};𝜇,𝚺)$
<br>
= log $\prod_{i=1}^{m}\frac{1}{(2𝜋)^{p/2 |Σ|^{1/2}}}exp(-\frac{1}{2}(x^{(i)} - \mu)^T𝚺^{-1}(x^{(i)}-\mu))$
<br>
= $\sum_{i=1}^{m}(-\frac{p}{2}log(2𝜋) - \frac{1}{2} log |Σ| -\frac{1}{2}(x^{(i)} - \mu)^T𝚺^{-1}(x^{(i)}-\mu))$
<br>
$𝑙(𝜇,Σ;)=−\frac{m𝑝}{2}log(2𝜋)−\frac{m}{2}log|Σ|−\frac{1}{2}\sum_{i=1}^{m}(x^{(i)} - \mu)^T𝚺^{-1}(x^{(i)}-\mu)$
<br>
<br>
### Deriving $\mu$ and $\Sigma$
<br>
To take the derivative with respect to 𝜇 and equate to zero we will make use of the following matrix calculus identity:
<br>
$\frac{∂}{∂𝜇}𝑙(𝜇,𝚺|𝐱^{(𝐢)}) = \sum_{i=1}^{m}𝚺^{−1}(𝜇−𝐱^{(𝐢)}) = 0$
<br>
<br>
$0 = 𝑚𝜇 − \sum_{i=1}^{m}𝐱^{(𝐢)}$
<br>
<br>
$\mu=\frac{1}{m}\sum_{i=1}^{m}𝐱^{(𝐢)}$
<br>
<br>
Deriving the MLE for the covariance matrix requires more work and the use of the following linear algebra and calculus properties:
<br>
<br>
Combining these properties allows us to calculate
<br>
<br>
$\frac{∂}{∂𝐴}𝑥^𝑡𝐴𝑥 = \frac{∂}{∂𝐴}𝑡𝑟[𝑥^𝑇𝑥𝐴]=[𝑥𝑥^𝑡]^𝑇=𝑥^{𝑇𝑇}𝑥^𝑇=𝑥𝑥^𝑇$
<br>
<br>
Which is the outer product of the vector 𝑥 with itself.
<br>
<br>
We can now re-write the log-likelihood function and compute the derivative w.r.t. $Σ^{−1}$ (note 𝐶 is constant)
<br>
<br>
$𝑙(𝜇,𝚺|𝐱^{(𝐢)}) = C − \frac{𝑚}{2}log|Σ| − \frac{1}{2}\sum_{𝑖}^{𝑚}(𝐱^{(𝐢)}−𝜇)^𝐓𝚺^{−1}(𝐱^{(𝐢)}−𝜇)$
<br>
<br>
= $C + \frac{𝑚}{2}log|Σ^{−1}|−\frac{1}{2}\sum_{𝑖}^{𝑚}𝑡𝑟[(𝐱^{(𝐢)}−𝜇)(𝐱^{(𝐢)}−𝜇)^𝐓𝚺^{−1}]$
<br>
<br>
= $\frac{∂}{∂Σ^{−1}}𝑙(𝜇,𝚺|𝐱^{(𝐢)})$
<br>
<br>
= $\frac{𝑚}{2}\sum - \frac{1}{2}\sum_{i=1}^{m}(X^{(i)} - \mu)(X^{(i)} - \mu)^T$
<br>
<br>
Since $𝚺^T = 𝚺$
<br>
<br>
$0 = m𝚺 - \sum_{i=1}^{m}(X^{(i)} - \mu)(X^{(i)} - \mu)^T$
<br>
<br>
$𝚺 = \frac{1}{m}\sum_{i=1}^{m}(X^{(i)} - \mu)(X^{(i)} - \mu)^T$

```python
# Sample Answer to (b)
import numpy as np # numpy
import math

def sge(X):
    """
    This function computes  the mean mu and standard deviation sigma 
    for a dataset coming from spherical Gaussian distribution N(mu, 2 sigma^2 I)

    @param X : A numpy array of size n x p where each row represents a p-dimensional data point
    e.g. X = [ [2, 1], [3, 7], [4 ,5] ] is a dataset with 3 samples having two coordinates each.
    @return mu: A numpy array of size p representing the estimated mean of the dataset
    @return sigma: A real number indicating the estimated standard deviation of the dataset
    """
    mu = np.zeros(X.shape[1])     # our vector of mean
    sigma = 0
    list1 = X[:,0]       # the first column in X 
    list2 = X[:,1]       # the second column in X 
    sum1 = 0
    sum2 = 0
    for i in range (len(list1)):
        sum1 += list1[i]         # the sum of the first column
    u1 = sum1 / len(list1)     # the mean of the first column
    
    for j in range (len(list2)):
        sum2 += list2[j]        # the sum of the second column
    u2 = sum2 / len(list2)   # the mean of the second column 
    mu = np.append(u1, u2)     # add two mean in mu
    
    total_sum1 = 0
    total_sum2 = 0
    for k in range (len(list1)):
        total_sum1 += (list1[k] - u1)**2     
    under_root1 = total_sum1 * 1.0 / len(list1)      # the variance of the first column
    
    for p in range (len(list2)):
        total_sum2 += (list2[p] - u2)**2
    under_root2 = total_sum2 * 1.0 / len(list2)    # the variance of the second column
    
    sigma =  math.sqrt(under_root1 + under_root2)    # the standard deviation of our data X

    return mu, sigma
```

```python
%matplotlib inline  
import numpy as np
import matplotlib
import matplotlib.pyplot as plt 

def myplot1(X):
    # Using only the first two features of the dataset X
    X = X[:, [0,1]]
    
    # Answer to (c) 1. A scatter plot of the original data X
    plt.plot( X[:,0], X[:,1], 'bo')
    
    # Answer to (c) 2. plot circles with center μ and radius r
    u1 = sge(X)[0][0]   # mean for the first column in X
    u2 = sge(X)[0][1]   # mean for the second column in X
    sigma = sge(X)[1]
    circle1=plt.Circle((u1, u2), math.sqrt(2)*1*sigma, color='r', fill=False)           # k = 1
    circle2=plt.Circle((u1, u2), math.sqrt(2)*2*sigma, color='blue', fill=False)     # k = 2
    circle3=plt.Circle((u1, u2), math.sqrt(2)*3*sigma, color='g', fill=False)         # k = 3
    
    # Legend for each circle, for our understanding, we need to legend the percentage of data outside the circle.
    counter1 = 0      # number of data in the first circle
    counter2 = 0     # number of data in the second circle
    counter3 = 0     # number of data in the third circle
    radius = math.sqrt(2)*1*sigma      

    for z in range (len(X[:,0])):
        dist = math.sqrt((X[z,0] - u1)**2 + ((X[z,1] - u2)**2))     # the distance between the point and the center point
        if radius - dist >= 0:      # check if the point is in the first circle
            counter1 +=1
        elif radius - dist <= 0 and radius*2 - dist >= 0: # check if the point is between first circle and second circle
            counter2 +=1  
        elif radius*2 - dist <= 0 and radius*3 - dist >= 0: # check if the point is between second circle and third circle
            counter3 +=1 

    in_circle1 = counter1 / len(X[:,0])      # float of data inside the first circle.
    in_circle2 = counter2 / len(X[:,0])    # float of data inside the second circle.
    in_circle3 = counter3 / len(X[:,0])    # float of data inside the third circle.
    out_circle1 = 1 - in_circle1              
    out_circle1 = "{0:.0%}".format(round(out_circle1, 2))     # percentage of data outside the first circle.
    out_circle2 = 1 - in_circle2 - in_circle1   
    out_circle2 = "{0:.0%}".format(round(out_circle2, 2))   # percentage of data outside the second circle.
    out_circle3 = 1 - in_circle3 - in_circle2 - in_circle1    
    out_circle3 = "{0:.0%}".format(round(out_circle3, 2))   # percentage of data outside the third circle.

    fig = plt.gcf()
    ax = fig.gca()
    ax.add_artist(circle1)
    ax.add_artist(circle2)
    ax.add_artist(circle3)
    ax.legend((circle1, circle2, circle3), (out_circle1, out_circle2, out_circle3))
    plt.axis([-200, 1200, -200, 1200])
    plt.show()
 
dataset_source = np.DataSource().open('http://bit.ly/HW1_2019 ')
X = np.loadtxt(dataset_source)

myplot1(X)
```
